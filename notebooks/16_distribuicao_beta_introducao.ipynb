# Introdu√ß√£o √† Distribui√ß√£o Beta

Este notebook apresenta a distribui√ß√£o Beta, sua intui√ß√£o, propriedades e um exemplo pr√°tico com simula√ß√£o em Python.

## Defini√ß√£o

A distribui√ß√£o Beta √© uma distribui√ß√£o cont√≠nua definida no intervalo [0,1]. Sua densidade √©:

$$ 
f(\theta; \alpha, \beta) = \frac{1}{B(\alpha, \beta)} \theta^{\alpha - 1} (1 - \theta)^{\beta - 1}
$$

- $\alpha$: par√¢metro relacionado a "sucessos"
- $\beta$: par√¢metro relacionado a "fracassos"
- $B(\alpha, \beta)$ √© a fun√ß√£o beta (uma constante de normaliza√ß√£o).


## Propriedades

- Suporte: $ \theta \in (0, 1) $
- M√©dia: $ \mathbb{E}[\theta] = \frac{\alpha}{\alpha + \beta} $
- Vari√¢ncia: $ \frac{\alpha \beta}{(\alpha + \beta)^2(\alpha + \beta + 1)} $

### Intui√ß√£o

* $\theta$ representa uma **propor√ß√£o** ou **probabilidade desconhecida**.
* $\alpha$: "n√∫mero de sucessos observados + 1" (em muitos casos pr√°ticos).
* $\beta$: "n√∫mero de fracassos observados + 1".

Quanto maiores $\alpha$ e $\beta$, mais concentrada √© a distribui√ß√£o.
Se $\alpha = \beta = 1$, temos a distribui√ß√£o uniforme no intervalo $[0,1]$.

---

### üéØ Exemplos Comuns

| Distribui√ß√£o Beta  | Forma da Distribui√ß√£o      | Significado Intuitivo                       |
| ------------------ | -------------------------- | ------------------------------------------- |
| $\text{Beta}(1,1)$ | Uniforme                   | Aus√™ncia de conhecimento pr√©vio             |
| $\text{Beta}(2,2)$ | Sim√©trica, centrada em 0.5 | Cren√ßa moderada de que $\theta \approx 0.5$ |
| $\text{Beta}(5,1)$ | Concentrada perto de 1     | Alta confian√ßa em $\theta$ grande           |
| $\text{Beta}(1,5)$ | Concentrada perto de 0     | Alta confian√ßa em $\theta$ pequeno          |

## Exemplo: Estimar probabilidade de sucesso

Suponha que observamos 8 sucessos e 2 fracassos. Vamos modelar $ \theta $ com uma Beta(9, 3).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Par√¢metros da distribui√ß√£o Beta
alpha_param = 8 + 1
beta_param = 2 + 1

# Valores de theta
theta = np.linspace(0, 1, 500)
pdf = beta.pdf(theta, alpha_param, beta_param)

# Gr√°fico
plt.plot(theta, pdf, label=f'Beta({alpha_param},{beta_param})')
plt.title('Distribui√ß√£o Beta para Œ∏')
plt.xlabel('Œ∏ (probabilidade de sucesso)')
plt.ylabel('Densidade')
plt.grid()
plt.legend()
plt.show()
```

## Estat√≠sticas da distribui√ß√£o

```python
mean = alpha_param / (alpha_param + beta_param)
var = (alpha_param * beta_param) / ((alpha_param + beta_param)**2 * (alpha_param + beta_param + 1))
print(f"M√©dia: {mean:.3f}")
print(f"Vari√¢ncia: {var:.4f}")
```
