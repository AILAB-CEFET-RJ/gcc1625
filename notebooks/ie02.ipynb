Exemplo: Podemos afirmar com certeza que a média populacional esteja realmente dentro desse intervalo? Não! Em vez disso, o que podemos afirmar é que, se repetíssemos o processo de amostragem uma grande quantidade de vezes e calculássemos o intervalo de confiança correspondente, em aproximadamente 95\% dos casos, a amostra conteria a média populacional.

Repare também que o intervalo para o nível 95\% ($\left[1,54; 1,62\right]$) é *maior* do que o intervalo para 68\% ($\left[1,56; 1,60\right]$). 

*  Como justificar o intervalo maior para 95\%?
*  O que significaria um intervalo de confiança no nível 100\%?


### Procedimento geral

O procedimento geral para a determinação de um intervalo de confiança para $\mu$, para uma amostra grande, é resumido a seguir.

Seja $X_{1}, X_{2}, \ldots, X_{n}$ uma amostra aleatória de tamanho $n$ ($n>30$) de uma população cuja média é $\mu$ e o desvio padrão é $\sigma$. Então um intervalo de confiança no nível $100(1-\alpha)$\% para $\mu$ é
	
$$
		\overline{X} \pm z_{\alpha/2}\sigma_{\overline{X}}
$$
em que $\sigma_{\overline{X}} = \sigma/\sqrt{n}$. 

Quando o valor de $\sigma$ é desconhecido, ele pode ser substituído pelo valor de $s$, a estimativa pontual do desvio padrão populacional.


# Intervalo de confiança para $p$

De forma análoga ao que foi feito para médias, há também procedimento geral para determinar intervalo de confiança para a proporção populacional ($p$).

Regra prática: o procedimento a seguir deve ser usado apenas quando a amostra contiver pelo menos 10 sucessos e 10 falhas.

Perceba a implicação prática da regra acima: 

*  Se a proporção populacional fosse próxima de 0,5, o tamanho da amostra necessário para produzir pelo menos 10 sucessos e pelo menos 10 falhas provavelmente seria próximo de 20. 
*  Mas se a proporção da população fosse extrema (i.e., próxima de 0 ou 1), provavelmente seria necessária uma amostra maior para produzir pelo menos 10 sucessos e 10 falhas. 
\end{itemize}
%https://stattrek.com/estimation/confidence-interval-proportion.aspx?tutorial=samp


Seja $X$ a variável aleatória que conta a quantidade de sucessos em uma amostra de tamanho $n$. Sendo assim, $X \sim  Bin(n,p)$, e a estimativa para $p$ (proporção populacional) é $\hat{p} = X/n$.

O desvio padrão da variável $\hat{p}$ é $\sqrt{p(1-p)/n}$.

Considerando que o tamanho da amostra é grande, pelo TLC, temos que:
$$
\hat{p} \sim  N\left(p, \frac{p(1-p)}{n}\right)
$$

Se considerarmos o nível de confiança de 95\%, então a proporção populacional satisfaz a seguinte desigualdade:
$$
\hat{p} - 1,96\sqrt{\frac{p(1-p)}{n}} < p < \hat{p} + 1,96\sqrt{\frac{p(1-p)}{n}}
$$

Repare que o intervalo de confiança acima depende de $p$! Na prática, para obter o intervalo de confiança para $p$ no nível $(1-\alpha)$\%, recomenda-se substituir $p$ por $\hat{p}$, na expressão acima. Sendo assim, obtemos o intervalo de confiança a seguir:
$$
\hat{p} \pm z_{\alpha/2}\sqrt{\frac{\hat{p}(1-\hat{p})}{n}}
$$

### Procedimento geral

Seja $\widehat{p}$ a proporção de sucessos em um grande número $n$ de ensaios independentes de Bernoulli com probabilidade de sucesso $p$. Então, o intervalo de confiança no nível $100(1-\alpha$)\% para $p$ é

$$
	\widehat{p}	\pm z_{\alpha/2}\sqrt{\frac{\widehat{p}(1-\widehat{p})}{n}}
$$
