```python
import numpy as np
import scipy.stats as stats
amostra = np.array([142, 138, 155, 149, 160, 137, 152, 145, 158, 143, 151, 147])
x_bar = np.mean(amostra)
s = np.std(amostra, ddof=1)
n = len(amostra)
t = stats.t.ppf(0.995, n-1)  # Valor crítico para um intervalo de confiança de 99%
margem_erro = t * (s / np.sqrt(n))
intervalo_inferior = x_bar - margem_erro
intervalo_superior = x_bar + margem_erro
print(f"Média amostral: {x_bar:.2f}")
print(f"Desvio padrão amostral: {s:.2f}")
print(f"Intervalo de confiança para a média: [{intervalo_inferior:.2f}, {intervalo_superior:.2f}]")
```
