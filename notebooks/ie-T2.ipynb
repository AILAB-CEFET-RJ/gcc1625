# 1 Teorema do Limite Central

```r
tcl.exp = function(n, N=1000, titulo=" ")
{
  medias=numeric(N)
  for (i in 1:N) 
      medias[i]= mean(rexp(n,.2))
    
  hist(medias, freq=F, main=titulo)
    
  x <- seq(0, 10, length=1000)
  y <- dnorm(x, 5, 5*sqrt(1/n) )
  points(x, y, type="l", lwd=2, col="red")
} 

tcl.exp(40,titulo="n=40")
```

# 2 Média e Variância Amostrais

### (i)

```r
mu <- 2
sigma <- 2
n <- 10
asim <- 10
xbar1 <- rep( NA , asim )
xvar1 <- rep( NA , asim )
for( i in 1: asim )
{
    set.seed ( i )
    #x contem uma amostra aleatoria de tamanho n da variavel X
    x <- rnorm(n , mu , sigma )
    xbar1 [ i ] <- mean( x )
    xvar1 [ i ] <- var ( x )
}
xbar1
xvar1
```

### (vi)

```r
mu<-2
sigma<-2

n<-10
asim<-1000

xbar<-rep( NA , asim )
xbar2<-rep( NA , asim )
xbar3<-rep( NA , asim )

xvar<-rep( NA , asim )
xvar2<-rep( NA , asim )
xvar3<-rep( NA , asim )

for( i in 1: asim )
{
    set.seed( i )
    x<-rnorm(n , mu , sigma )
    x2<-rnorm( n*10 , mu , sigma )
    x3<-rnorm( n*100 , mu , sigma )
    xbar [ i ]<-mean( x )
    xbar2 [ i ]<-mean( x2 )
    xbar3 [ i ]<-mean( x3 )
    xvar [ i ]<-var( x )
    xvar2 [ i ]<-var( x2 )
    xvar3 [ i ]<-var( x3 )
}

mean( xbar )
mean( xbar2 )
mean( xbar3 )
var( xbar )
var( xbar2 )
var( xbar3 )
mean( xvar )
mean( xvar2 )
mean( xvar3 )
```

- A média das medias amostrais é sempre próxima da média populacional
- A vaiância das médias amostrais é aproximadamente singma^2/n
- A media das variancias amostrais é aproximadamente sigma^2
