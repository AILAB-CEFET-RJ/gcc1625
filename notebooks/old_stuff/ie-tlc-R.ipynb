```r
n = 300
repeticoes = 10000
lambda = 0.2
amostras <- replicate(repeticoes, rexp(n, rate = lambda))
amostras.media <- colMeans(amostras)
```

```r
set.seed(1234)
df <- data.frame(
  medias=amostras.media
  )
head(df)
```

```r
library(ggplot2)
# Basic histogram
ggplot(df, aes(x=medias)) + geom_histogram()
# Change the width of bins
ggplot(df, aes(x=medias)) + 
  geom_histogram(binwidth=1)
# Change colors
p<-ggplot(df, aes(x=medias)) + 
  geom_histogram(color="black", fill="white")
p
```
