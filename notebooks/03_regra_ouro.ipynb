O código a seguir exibe três subplots, cada um destacando um intervalo específico: 68%, 95%, e 99.7%, correspondendo a 1, 2, e 3 desvios padrão ao redor da média de uma variável aleatória normal padrão.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parâmetros da distribuição normal
mu = 0     # média
sigma = 1  # desvio padrão

# Gerando valores para o eixo x
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# Criando subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Subplot 1: 68% (1 desvio padrão)
axs[0].plot(x, y, color='blue')
axs[0].fill_between(x, y, where=(x >= mu - sigma) & (x <= mu + sigma), color='yellow', alpha=0.5)
axs[0].set_title("68% dos Dados (1σ)")
axs[0].set_xlabel("Valores")
axs[0].set_ylabel("Densidade de Probabilidade")
axs[0].grid(True)

# Subplot 2: 95% (2 desvios padrão)
axs[1].plot(x, y, color='blue')
axs[1].fill_between(x, y, where=(x >= mu - 2*sigma) & (x <= mu + 2*sigma), color='orange', alpha=0.5)
axs[1].set_title("95% dos Dados (2σ)")
axs[1].set_xlabel("Valores")
axs[1].grid(True)

# Subplot 3: 99.7% (3 desvios padrão)
axs[2].plot(x, y, color='blue')
axs[2].fill_between(x, y, where=(x >= mu - 3*sigma) & (x <= mu + 3*sigma), color='red', alpha=0.5)
axs[2].set_title("99.7% dos Dados (3σ)")
axs[2].set_xlabel("Valores")
axs[2].grid(True)

# Configurações gerais
plt.suptitle("Regra 68-95-99.7% em uma Distribuição Normal")
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

O código abaixo simula a regra 68-95-99.7% usando uma distribuição normal com média 10 e variância 2. O código gera dados simulados e calcula a porcentagem de valores que caem dentro de 1, 2 e 3 desvios padrão da média, confirmando experimentalmente a regra.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parâmetros da distribuição normal
mu = 10       # média
sigma = np.sqrt(2)  # desvio padrão (sqrt da variância)

# Gerar dados simulados
np.random.seed(42)
data = np.random.normal(mu, sigma, 10000)  # 10,000 amostras

# Definindo os intervalos de 1σ, 2σ e 3σ
within_1_sigma = np.mean((data >= mu - sigma) & (data <= mu + sigma)) * 100
within_2_sigma = np.mean((data >= mu - 2*sigma) & (data <= mu + 2*sigma)) * 100
within_3_sigma = np.mean((data >= mu - 3*sigma) & (data <= mu + 3*sigma)) * 100

# Exibindo os resultados
print("Simulação da regra 68-95-99.7% para uma distribuição normal (µ=10, σ²=2):")
print(f"Percentual de dados dentro de 1σ (68% esperado): {within_1_sigma:.2f}%")
print(f"Percentual de dados dentro de 2σ (95% esperado): {within_2_sigma:.2f}%")
print(f"Percentual de dados dentro de 3σ (99.7% esperado): {within_3_sigma:.2f}%")
```
