# 02 Distribuições Amostrais - Prática em Python

Este notebook foca na parte operacional: gerar amostras, calcular estatísticas e visualizar distribuições amostrais.

## Objetivo da prática

Comparar, por simulação, a distribuição da população, a distribuição de uma amostra e a distribuição amostral da média.

## Fluxo da atividade

1. Simulação com população normal e médias amostrais.
2. Simulação com lançamentos de dado.
3. Construção de distribuição amostral empírica de `x̄`.

## Distribuição amostral da média

Nesta prática, a estatística de interesse é a média amostral `x̄`, calculada repetidamente sobre amostras de tamanho fixo.

## Simulação computacional 1

Vamos gerar uma população sintética normal e, em seguida, várias amostras para montar a distribuição amostral das médias.

O código abaixo gera:

- a população simulada;
- múltiplas médias amostrais;
- histogramas para comparação visual;
- uma checagem numérica simples da média e variância das médias amostrais.

Para reprodutibilidade, usamos semente fixa.

```python
import numpy as np
import matplotlib.pyplot as plt

RANDOM_STATE = 42
rng = np.random.default_rng(RANDOM_STATE)

# parâmetros da população normal
mu = 0
sigma = 1
populacao_tamanho = 10_000
amostra_tamanho = 30
num_amostras = 1_000

# população sintética
populacao = rng.normal(loc=mu, scale=sigma, size=populacao_tamanho)

# distribuição amostral empírica da média (amostras com reposição)
medias_amostrais = np.array([
    rng.choice(populacao, size=amostra_tamanho, replace=True).mean()
    for _ in range(num_amostras)
])

# checagem rápida: E[X̄]≈mu e Var[X̄]≈sigma^2/n
print(f"Média empírica de X̄: {medias_amostrais.mean():.4f} (teórico: {mu:.4f})")
print(f"Variância empírica de X̄: {medias_amostrais.var(ddof=1):.4f} (teórico: {sigma**2/amostra_tamanho:.4f})")

# visualização
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(populacao, bins=30, alpha=0.75, color='steelblue', edgecolor='black')
plt.title('População simulada (Normal)')
plt.xlabel('Valor')
plt.ylabel('Frequência')

plt.subplot(1, 2, 2)
plt.hist(medias_amostrais, bins=30, alpha=0.75, color='darkorange', edgecolor='black')
plt.title('Distribuição amostral empírica de X̄')
plt.xlabel('Média da amostra')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```

## Simulação computacional 2 (dado de 6 lados)

Agora usamos lançamentos de dado para repetir a lógica de população, amostra e distribuição amostral.

Para um dado justo: `μ = 3.5` e `σ² ≈ 2.9166`, valores úteis para comparar com os resultados empíricos.

Primeiro, geramos uma amostra pequena de lançamentos para inspecionar valores e frequências.

```python
# amostra pequena de lançamentos do dado
n_pequeno = 50
rolls = rng.integers(1, 7, size=n_pequeno)

labels, counts = np.unique(rolls, return_counts=True)
print('Primeiras observações:', rolls[:20])
print('Contagens por face:', dict(zip(labels, counts)))
```

Agora geramos uma amostra grande e visualizamos o histograma da distribuição da amostra.

```python
# amostra grande para aproximar a distribuição da população do dado
sample = rng.integers(1, 7, size=10_000)
labels, counts = np.unique(sample, return_counts=True)
proportions = counts / counts.sum()

plt.figure(figsize=(6, 4))
plt.bar(labels, proportions, align='center')
plt.axhline(1/6, color='red', linestyle='--', label='Proporção teórica (1/6)')
plt.xticks(labels)
plt.ylim(0, 0.22)
plt.title('Distribuição empírica dos resultados do dado')
plt.xlabel('Face')
plt.ylabel('Proporção')
plt.legend()
plt.show()
```

Para um tamanho amostral fixo `n`, calculamos uma média amostral `x̄`.

```python
# uma média amostral para n fixo
n = 50
rolls = rng.integers(1, 7, size=n)
xbar = rolls.mean()

print('Média amostral (X̄):', round(xbar, 4))
print('Média teórica da população do dado (μ): 3.5')
print('Erro absoluto |X̄ - μ|:', round(abs(xbar - 3.5), 4))
```

```python
# referência teórica para a variância do dado justo
# Var(X) = 35/12 e Var(X̄) = Var(X)/n
var_pop = 35 / 12
print('Var(X) teórica:', round(var_pop, 6))
print('Var(X̄) teórica para n=50:', round(var_pop / 50, 6))
```

Repetindo o processo muitas vezes, construímos a distribuição amostral empírica de `x̄`.

```python
# distribuição amostral empírica de X̄ para diferentes n
num_amostras = 100_000
n = 5
means = np.array([rng.integers(1, 7, size=n).mean() for _ in range(num_amostras)])

mean_emp = means.mean()
var_emp = means.var(ddof=1)
mean_theo = 3.5
var_theo = (35/12) / n

print(f"E[X̄] empírica: {mean_emp:.4f} (teórica: {mean_theo:.4f})")
print(f"Var(X̄) empírica: {var_emp:.4f} (teórica: {var_theo:.4f})")

plt.figure(figsize=(7,4))
plt.hist(means, bins=80, density=True, alpha=0.8, color='slateblue', edgecolor='black')
plt.title(f'Distribuição amostral empírica de X̄ (n={n})')
plt.xlabel('X̄')
plt.ylabel('Densidade')
plt.show()
```
